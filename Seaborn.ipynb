# Seaborn Tutorial

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

## tip is the dependent variabel and other variable is independent

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr()) 
```

# heatmap shows the correlation between the variables

# Join plot
### A join plot allows to study the relationship between 2 numberic variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

## Pair Plot
A pair plot is also knows as scatter plot, in which one variable in the same data row is match with another variables value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
df["smoker"].value_counts()
```

```python
sns.pairplot(df,hue="smoker")
```

## Dist Plot
### Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df["tip"])
```

```python
sns.displot(df["tip"],kde=False,bins=10) # kde means kernel density 
```

## Categorical Plots 
## seaborn also helps us in doing the analysis on Categorial Data points. In this section we will discuss about

    Boxplot
    violinplot countplot
    countplot
    bar plot


```python
sns.countplot("sex",data=df) # Show the counts of observations in each categorical bin using bars.
```

```python
sns.countplot("day",data=df)
```

```python
sns.countplot(y="sex",data=df)
```

```python
# Bar plot
sns.barplot(y="total_bill",x="smoker",data=df)
```

## Box Plot
### A box and whisker plot (sometimes call a boxplot) is a graph that presents information from a five- number summary.

```python
sns.boxplot(x="day",y="total_bill",data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,orient="v")
```

```python
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df) # hue means classifing this data with respect to smoker
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill",y="day",data=df,palette="rainbow")
```
