## Matplotlib
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK

## Some of the major Pros of Matplotlib are:
 Generally easy to get started for simple plots 
 Support for custom labels and texts
Great control of every element in a figure 
 High-quality output in many formats 
 very coustomizable in general
  
 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c="g")
plt.xlabel("x axies")
plt.ylabel("y axies")
plt.title("2D tabel")
```

```python
y=x*x
```

```python
plt.plot(x,y,"r",linestyle="dashed",linewidth=2, markersize=12)
```

## Creating Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,"r")
plt.subplot(2,2,2)
plt.plot(x,y,"g")
plt.subplot(2,2,3)
plt.plot(x,y,"b")
```

# Compute the x and  coordinates for points on a sine curve

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# Make first plot
plt.plot(x,y_sin)
plt.title("Sine")

# Make the second subplot as active and made the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,"g")
plt.title("Cosine")
# Showing the figure.
plt.show()
```

# # Bar plot 

```python
# Bar plot 
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color="g")
plt.title("Bar graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```

# Histograms

```python
a=np.array([22,87,5,43,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# Box Plot using Matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
data
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```

# Pie Chart

```python
# Data to plot
labels="Python","C++","Ruby","java"
sizes=[215,130,245,210]
colors=["gold","yellowgreen","lightcoral","lightskyblue"]
explode=[0.1,0.1,0,0] # explode 1st slice
# Plot 
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True) 
# "%1.1f%%" this means autopercent will be shown as 1.1 floating
```
